```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
1Ô∏è‚É£ Load the dataset
df = pd.read_csv("Coffe_sales.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

print("‚úÖ Data Loaded Successfully")
print(df.head())
```

```python
 # Total Sales Revenue
total_sales = df['money'].sum()
print("üí∞ Total Sales Revenue:", round(total_sales, 2))

plt.figure(figsize=(4,4))
plt.bar(['Total Revenue'], [total_sales], color='orange')
plt.title('Total Sales Revenue')
plt.ylabel('Revenue (‚Çπ)')
plt.show()

```

```python
# Most Popular Coffee
popular_coffee = df['coffee_name'].value_counts().head(1)
print("‚òï Most Popular Coffee:\n", popular_coffee)

plt.figure(figsize=(5,4))
sns.barplot(x=popular_coffee.index, y=popular_coffee.values, palette='magma')
plt.title('Most Popular Coffee')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Top 3 Best-Selling Coffee Types
top3 = df.groupby('coffee_name')['money'].sum().sort_values(ascending=False).head(3)
print("üèÜ Top 3 Best-Selling Coffees:\n", top3)

plt.figure(figsize=(6,4))
sns.barplot(x=top3.index, y=top3.values, palette='viridis')
plt.title('Top 3 Best-Selling Coffees')
plt.ylabel('Total Sales (‚Çπ)')
plt.xlabel('Coffee Type')
plt.show()
```

```python
avg_sale = df['money'].mean()
print("üìà Average Sale Value:", round(avg_sale, 2))

plt.figure(figsize=(4,4))
plt.bar(['Average Sale'], [avg_sale], color='green')
plt.title('Average Sale Value')
plt.ylabel('‚Çπ per Sale')
plt.show()
```

```python
week_sales = df.groupby('Weeks')['money'].sum().sort_values(ascending=False)
print("üìÖ Sales by Weekday:\n", week_sales)

plt.figure(figsize=(8,4))
sns.barplot(x=week_sales.index, y=week_sales.values, palette='coolwarm')
plt.title('Weekend vs Weekday Sales')
plt.ylabel('Total Sales (‚Çπ)')
plt.xlabel('Day of Week')
plt.show()
```

```python
# Most Profitable Month
month_sales = df.groupby('Month_name')['money'].sum().sort_values(ascending=False)
print("üìÜ Monthly Sales:\n", month_sales)

plt.figure(figsize=(8,4))
sns.barplot(x=month_sales.index, y=month_sales.values, palette='plasma')
plt.title('Most Profitable Month')
plt.ylabel('Total Sales (‚Çπ)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
hour_sales = df.groupby('Hours')['money'].sum().sort_values(ascending=False)
print("‚è∞ Hourly Sales:\n", hour_sales)

plt.figure(figsize=(8,4))
sns.lineplot(x=hour_sales.index, y=hour_sales.values, marker='o', color='teal')
plt.title('Peak Hour of Sales')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales (‚Çπ)')
plt.show()
```

```python
daily_sales = df.groupby('Date')['money'].sum().sort_values(ascending=False)
top_day = daily_sales.head(1)
print("üåü Day with Highest Total Sales:\n", top_day)

plt.figure(figsize=(10,4))
sns.lineplot(x=daily_sales.index, y=daily_sales.values, color='red')
plt.title('Total Sales by Date')
plt.xlabel('Date')
plt.ylabel('Total Sales (‚Çπ)')
plt.show()

print("‚úÖ Visualization Completed Successfully!")
```
